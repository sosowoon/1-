```
import requests
import pandas as pd
```

```
from bs4 import BeautifulSoup
```

```
url = "https://movie.naver.com/movie/running/current.nhn"
result = requests.get(url)
result
```

```
content = BeautifulSoup(result.content,"html.parser")
```

```
titles_container = content.findAll('dt',{'class':'tit'})
titles_container
```

```
title = []
for x in titles_container:
  title.append(x.find('a').text)

title
```

```
len(titles_container)
```

```
# 컬럼에 들어가는 값은 모두 같은 형으로 이것을 벡터라고 한다.
scores = content.select('dl.info_star dd div.star_t1 a span.num')
len(scores)
```

```
scores_list = []
for x in scores:
  scores_list.append(float(x.string))
scores_list
```

```
df = pd.DataFrame({"영화이름":title, "평점": scores_list})
df.head()
```
